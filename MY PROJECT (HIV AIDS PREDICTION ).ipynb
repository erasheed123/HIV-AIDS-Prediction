```python
x
```

```python
#to change the feature data column with string to integer values
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
x["Marital Staus"]=le.fit_transform(x["Marital Staus"])
x["STD"]=le.fit_transform(x["STD"])
x["Educational Background"]=le.fit_transform(x["Educational Background"])
x["HIV TEST IN PAST YEAR"]=le.fit_transform(x["HIV TEST IN PAST YEAR"])
x["AIDS education"]=le.fit_transform(x["AIDS education"])
x["Places of seeking sex partners"]=le.fit_transform(x["Places of seeking sex partners"])
x["SEXUAL ORIENTATION"]=le.fit_transform(x["SEXUAL ORIENTATION"])
x["Drug- taking"]=le.fit_transform(x["Drug- taking"])
x["Result"]=le.fit_transform(x["Result"])
x
```

```python
#Converting into array format
```

```python
x.values
```

```python
#Extracting indepedent and dependent variables
```

```python
#locating Feature values
x1=x.iloc[:,:9].values
x1
```

```python
#locating target values
y=x.iloc[:,-1].values
y
```

# LOGISTIC REGRESSION(CLASSIFICATION)

```python
# Splitting the dataset into training and test set.  
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test= train_test_split(x1, y, test_size= 0.2, random_state=100)  
```

```python
#TRAINING RESULTS OF X VALUE IN ARRAY FORMAT
x_train
```

```python
#TESTING RESULTS OF X VALUES IN AN ARRAY FORMAT
x_test
```

```python
#TRAINING RESULTS OF Y VALUES IN AN ARRAY FORMAT
y_train
```

```python
#TESTING RESULTS OF Y VALUES IN AN ARRAY FORMAT
y_test
```

```python
#Feature and target scaling(Standard scalar and min-max scalar) for preprocessing
from sklearn.preprocessing import MinMaxScaler
mmx=MinMaxScaler()
x_train=mmx.fit_transform(x_train)
x_test=mmx.transform(x_test)
```

```python
#Fitting the lR(Classification) model to the training set:
from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression()
print ( "REGRESSION MODEL : ", log_reg . fit(x_train , y_train) )
```

```python
#Predicting the Train set result; 
x_pred=log_reg.predict(x_train)
x_pred   
```

```python
#Predicting the Test set result;  
y_pred=log_reg.predict(x_test)  
y_pred
```

```python
#ACCURACY SCORES REGARDING LOGISTIC REGRESSION(CLASSIFICATION)
logistic_train_acc=log_reg . score (x_train , y_train)
logistic_test_acc=log_reg . score (x_test , y_test)

print ( "TRAINING DATA ACCURACY : " , logistic_train_acc) # training data accuracy
print ( "TESTING DATA ACCURACY : " , logistic_test_acc) # testing data accuracy
```

```python
#ACCURACY TEST SCORES REGARDING LOGISTIC REGRESSION(CLASSIFICATION)
q=log_reg.score(x_test,y_test)
q
```

# OUTPUT RESULT PREDICTION FOR LOGISTIC REGRESSION

```python
inputs=[[ 19,  2,  0,  4,  0,  0,  4,  1,  0]]

Result=log_reg.predict(inputs)

print(Result)

if(Result==0):
    print("HIV NEGATIVE")
else:
    print("HIV POSITIVE!!")
```
